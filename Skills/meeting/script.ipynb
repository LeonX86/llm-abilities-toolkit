```python
# 读取文件
with open('./data/opc会议转写.md', 'r', encoding='utf-8') as f:
    content = f.read()

# 整体清洗：删除所有 "发言人 X XX:XX" 格式的脏数据行
cleaned_lines = []
for line in content.split('\n'):
    parts = line.strip().split()
    # 判断是否为脏数据：发言人 + 数字 + 时间格式
    if len(parts) == 3 and parts[0] == '发言人' and parts[1].isdigit() and ':' in parts[2]:
        continue
    cleaned_lines.append(line)

cleaned_content = '\n'.join(cleaned_lines)

# 分割基本信息和会议转写
parts = cleaned_content.split('## 会议转写')
meeting_info = parts[0].strip()

# 按 ### 切分发言段落
speeches = []
if len(parts) > 1:
    sections = parts[1].split('###')
    for section in sections[1:]:  # 跳过第一个空段
        section = section.strip()
        if section:
            speeches.append('### ' + section)

# 输出结果
print("=" * 50)
print(meeting_info)
print("\n" + "=" * 50)
print(f"\n共有 {len(speeches)} 段发言\n")
for idx, speech in enumerate(speeches, 1):
    print(f"第 {idx} 段:")
    print(speech[:100] + "..." if len(speech) > 100 else speech)
    print("-" * 50)
```
